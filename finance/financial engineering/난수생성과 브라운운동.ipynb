dX = 정규분포를 통해 뽑아낸 난수 X 시간 프레임에 따른 스케일링 팩터

```python
import math
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Define dt, dqrt_dt

T = 1 # 1년
N = 10000 # 시행횟수
dt = T / N
sqrt_dt = math.sqrt(dt) # dt의 제곱근
```

```python
# Generate Series of Time, White Noise & Brownian Motin

t = np.array([i for i in np.arange(0, T, T/N)]) # 시간 배열(현재 시점부터 1년가지 dt만큼 증가하는 수열)
dX = np.random.randn(N) * sqrt_dt # 총 N개의 난수를 뽑아 배열 할당 후 스케일링팩터 곱함 
X = np.cumsum(dX) # dX를 시간이 지남에 따라 더하기
```

```python
# Print Graph

plt.subplot(2,1,1)
plt.plot(t,dX) # 백색 소음 시계열(dX)
plt.subplot(2,1,2)
plt.plot(t,X) # 백색 소음의 누적합 시계열(X)
```
