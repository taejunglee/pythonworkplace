```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
S0 = 100     # 초기값
mu = 0.1     # 추세파라미터
sigma = 0.23 # 변동성
T = 1.0      # 만기
D = 252      # 거래일
dt = T/D
N = 10000
```

```python
S = np.zeros((D+1, N))
S[0] = S0
```

```python
for t in range(1, D+1):
    S[t] = S[t-1]*np.exp((mu-0.5*sigma**2) * dt + sigma * np.sqrt(dt) * np.random.randn(N))
```

```python
plt.figure(figsize=(18,10))
plt.plot(S[:,:10000]) # [날짜범위, 시나리오 범위]
plt.xlabel('day')
plt.ylabel('price')
plt.grid(True)
plt.show()
```
